```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =============================================================
# ðŸ“‚ PATH CONFIG
# =============================================================
base_path = r"C:\Users\NXTWAVE\Downloads\Smart Solar-Powered Irrigation & Soil Optimization System"
data_path = os.path.join(base_path, "archive", "spg.csv")
model_path = os.path.join(base_path, "agrovolt_model.h5")
scaler_path = os.path.join(base_path, "agrovolt_scaler.pkl")
result_path = os.path.join(base_path, "agrovolt_result.csv")
visual_dir = os.path.join(base_path, "visuals")
os.makedirs(visual_dir, exist_ok=True)

# =============================================================
# ðŸ“¥ LOAD MODEL & DATA
# =============================================================
print("[INFO] Loading model, data and scaler...")
model = load_model(model_path)
scaler = joblib.load(scaler_path)
df = pd.read_csv(data_path).dropna()

print(f"[INFO] Dataset shape: {df.shape}")

# Assuming last column is the target
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
X_scaled = scaler.transform(X)
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

# =============================================================
# ðŸ”® PREDICTIONS
# =============================================================
print("[INFO] Generating predictions...")
y_pred = model.predict(X_scaled)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("\nðŸ“Š Model Performance Metrics:")
print(f"MAE : {mae:.4f}")
print(f"MSE : {mse:.4f}")
print(f"R2  : {r2:.4f}")

# Save results as CSV
result_df = pd.DataFrame({
    "Actual": y,
    "Predicted": y_pred.flatten()
})
result_df.to_csv(result_path, index=False)
print(f"[INFO] Results saved to: {result_path}")

# =============================================================
# ðŸ“ˆ ACCURACY GRAPH (from training history if available)
# =============================================================
history_file = os.path.join(base_path, "agrovolt_history.json")

if os.path.exists(history_file):
    with open(history_file, "r") as f:
        history = json.load(f)
    plt.figure(figsize=(8, 6))
    plt.plot(history["loss"], label='Train Loss')
    plt.plot(history["val_loss"], label='Val Loss')
    plt.legend()
    plt.title("AgroVolt Training Accuracy Graph")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.show()
    plt.savefig(os.path.join(visual_dir, "agrovolt_accuracy_graph.png"))
else:
    print("[WARN] No history file found â€” skipping accuracy graph.")

# =============================================================
# ðŸ”¥ HEATMAP OF FEATURE CORRELATIONS
# =============================================================
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("AgroVolt Feature Correlation Heatmap")
plt.show()
plt.savefig(os.path.join(visual_dir, "agrovolt_heatmap.png"))

# =============================================================
# ðŸ“Š PREDICTION VS ACTUAL (Scatter Plot)
# =============================================================
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, alpha=0.6, color='green')
plt.title("AgroVolt â€” Predicted vs Actual Output")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.grid(True)
plt.show()
plt.savefig(os.path.join(visual_dir, "agrovolt_comparison_graph.png"))

# =============================================================
# ðŸ“‰ RESULT GRAPH (Line Plot)
# =============================================================
plt.figure(figsize=(10, 6))
plt.plot(y[:200], label="Actual", linewidth=2)
plt.plot(y_pred[:200], label="Predicted", linestyle='dashed')
plt.title("AgroVolt â€” Result Graph (First 200 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Generated Power / Target Variable")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig(os.path.join(visual_dir, "agrovolt_result_graph.png"))

# =============================================================
# ðŸ’¾ SAVE PERFORMANCE METRICS
# =============================================================
metrics = {
    "MAE": float(mae),
    "MSE": float(mse),
    "R2": float(r2)
}
json_path = os.path.join(base_path, "agrovolt_prediction.json")
yaml_path = os.path.join(base_path, "agrovolt_config.yaml")

with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)
with open(yaml_path, "w") as f:
    yaml.dump(metrics, f)

print(f"[INFO] Metrics saved as JSON and YAML.")
print(f"[INFO] All visuals stored in: {visual_dir}")
print("âœ… Visualization complete. All graphs displayed and saved successfully!")
```
