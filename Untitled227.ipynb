```python
# =============================================================
# ðŸŒž AGROVOLT â€” Hybrid AIS + CSA Optimization with CNN-LSTM
# =============================================================

import os
import json
import yaml
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Dropout
from tensorflow.keras.optimizers import Adam

# =============================================================
# ðŸ“‚ PATHS
# =============================================================
base_path = r"C:\Users\NXTWAVE\Downloads\Smart Solar-Powered Irrigation & Soil Optimization System"
data_path = os.path.join(base_path, "archive", "spg.csv")
visual_dir = os.path.join(base_path, "visuals")
os.makedirs(visual_dir, exist_ok=True)

# Output files with hybrid_ prefix
model_path = os.path.join(base_path, "hybrid_agrovolt_model.h5")
scaler_path = os.path.join(base_path, "hybrid_agrovolt_scaler.pkl")
config_path = os.path.join(base_path, "hybrid_agrovolt_config.yaml")
json_path = os.path.join(base_path, "hybrid_agrovolt_prediction.json")
result_path = os.path.join(base_path, "hybrid_agrovolt_result.csv")

# =============================================================
# ðŸ“¥ Load Dataset
# =============================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(data_path)
df = df.dropna()
print(f"[INFO] Shape: {df.shape}")

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# =============================================================
# ðŸ§® Data Scaling
# =============================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, scaler_path)
print(f"[INFO] Scaler saved at: {scaler_path}")

# Reshape for CNN-LSTM
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# =============================================================
# âš™ï¸ Hybrid AIS + CSA Optimizer (with early stopping)
# =============================================================

def fitness(params):
    """Train CNN-LSTM with given params and return MSE."""
    lr, neurons, dropout = params
    model = Sequential([
        Conv1D(32, 2, activation='relu', input_shape=(X_train.shape[1], 1)),
        MaxPooling1D(),
        LSTM(int(neurons), activation='tanh', return_sequences=False),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
    model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)
    pred = model.predict(X_test)
    return mean_squared_error(y_test, pred)


def hybrid_ais_csa(iterations=10, population_size=6, early_stop=3):
    """Hybrid AIS + CSA optimizer with early stopping."""
    lb = np.array([0.0005, 16, 0.1])  # learning rate, neurons, dropout
    ub = np.array([0.01, 128, 0.5])
    dim = 3

    population = np.random.uniform(lb, ub, (population_size, dim))
    best_score = float("inf")
    best_sol = None
    no_improve = 0

    awareness_prob = 0.1   # CSA parameter
    flight_length = 2.0    # CSA parameter

    for t in range(iterations):
        # Evaluate fitness
        fitness_vals = np.array([fitness(p) for p in population])
        sorted_idx = np.argsort(fitness_vals)
        mem_count = max(2, population_size // 2)
        memory = population[sorted_idx[:mem_count]]

        if fitness_vals[sorted_idx[0]] < best_score:
            best_score = fitness_vals[sorted_idx[0]]
            best_sol = population[sorted_idx[0]].copy()
            no_improve = 0
        else:
            no_improve += 1

        # Early stopping condition
        if no_improve >= early_stop:
            print(f"[INFO] Early stopping at iteration {t+1} (no improvement for {early_stop} rounds)")
            break

        # CSA + AIS update rules
        for i in range(population_size):
            r = np.random.rand()
            if r >= awareness_prob:
                j = np.random.randint(0, memory.shape[0])  # Safe index
                population[i] = population[i] + np.random.rand() * flight_length * (memory[j] - population[i])
            else:
                population[i] = np.random.uniform(lb, ub)

            # AIS clonal mutation
            mutation = np.random.normal(0, 0.05, dim)
            population[i] += mutation
            population[i] = np.clip(population[i], lb, ub)

        print(f"[Iter {t+1}/{iterations}] Best MSE so far: {best_score:.6f}")

    return best_sol, best_score


# =============================================================
# ðŸ§  Run Optimization + Final Model
# =============================================================
best_params, best_score = hybrid_ais_csa(iterations=8, population_size=5, early_stop=2)
lr, neurons, dropout = best_params
print(f"\n[INFO] âœ… Optimized Params -> lr={lr:.5f}, neurons={int(neurons)}, dropout={dropout:.2f}\n")

# =============================================================
# ðŸ§  Build Final CNN-LSTM Model
# =============================================================
model = Sequential([
    Conv1D(32, 2, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(),
    LSTM(int(neurons), activation='tanh', return_sequences=False),
    Dropout(dropout),
    Dense(1)
])
model.compile(optimizer=Adam(learning_rate=lr), loss='mse', metrics=['mae'])

history = model.fit(X_train, y_train, epochs=50, batch_size=32,
                    validation_split=0.2, verbose=1)

model.save(model_path)
print(f"[INFO] Model saved at: {model_path}")

# =============================================================
# ðŸ“ˆ Predictions & Evaluation
# =============================================================
pred = model.predict(X_test)
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred)

metrics = {
    "MAE": float(mae),
    "MSE": float(mse),
    "R2": float(r2),
    "Best_Params": {
        "Learning_Rate": float(lr),
        "Neurons": int(neurons),
        "Dropout": float(dropout)
    }
}

# Save metrics
with open(json_path, "w") as f: json.dump(metrics, f, indent=4)
with open(config_path, "w") as f: yaml.dump(metrics, f)
print(f"[INFO] Metrics saved to JSON and YAML.")

# Save results
result_df = pd.DataFrame({"Actual": y_test, "Predicted": pred.flatten()})
result_df.to_csv(result_path, index=False)
print(f"[INFO] Result CSV saved to: {result_path}")

# =============================================================
# ðŸ“Š Visualization Section
# =============================================================

# 1ï¸âƒ£ Accuracy Graph
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Hybrid AIS+CSA AgroVolt Accuracy Graph")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(visual_dir, "hybrid_agrovolt_accuracy_graph.png"))
plt.show()

# 2ï¸âƒ£ Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Hybrid AgroVolt Feature Correlation Heatmap")
plt.savefig(os.path.join(visual_dir, "hybrid_agrovolt_heatmap.png"))
plt.show()

# 3ï¸âƒ£ Comparison Scatter
plt.figure(figsize=(8,6))
plt.scatter(y_test, pred, color='green', alpha=0.6)
plt.title("Hybrid AgroVolt Predicted vs Actual")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.grid(True)
plt.savefig(os.path.join(visual_dir, "hybrid_agrovolt_comparison_graph.png"))
plt.show()

# 4ï¸âƒ£ Result Line Graph
plt.figure(figsize=(10,6))
plt.plot(y_test[:200], label="Actual", linewidth=2)
plt.plot(pred[:200], label="Predicted", linestyle='dashed')
plt.title("Hybrid AgroVolt Result Graph (First 200 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Target Value")
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(visual_dir, "hybrid_agrovolt_result_graph.png"))
plt.show()

print("\nâœ… HYBRID AIS + CSA AGROVOLT MODEL EXECUTED SUCCESSFULLY âœ…")
print(f"[INFO] All outputs saved in: {base_path}")
```
