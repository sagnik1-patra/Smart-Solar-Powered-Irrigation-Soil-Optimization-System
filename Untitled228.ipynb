```python
import os
import json
import yaml
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Dropout
from tensorflow.keras.optimizers import Adam

# =============================================================
# ðŸ“‚ PATHS
# =============================================================
base_path = r"C:\Users\NXTWAVE\Downloads\Smart Solar-Powered Irrigation & Soil Optimization System"
data_path = os.path.join(base_path, "archive", "spg.csv")
visual_dir = os.path.join(base_path, "visuals")
os.makedirs(visual_dir, exist_ok=True)

# Output files (ais_pso_ prefix)
model_path  = os.path.join(base_path, "ais_pso_agrovolt_model.h5")
scaler_path = os.path.join(base_path, "ais_pso_agrovolt_scaler.pkl")
config_path = os.path.join(base_path, "ais_pso_agrovolt_config.yaml")
json_path   = os.path.join(base_path, "ais_pso_agrovolt_prediction.json")
result_path = os.path.join(base_path, "ais_pso_agrovolt_result.csv")

# =============================================================
# ðŸ“¥ Load Dataset
# =============================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(data_path).dropna()
print(f"[INFO] Shape: {df.shape}")

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# =============================================================
# ðŸ§® Scaling
# =============================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, scaler_path)
print(f"[INFO] Scaler saved at: {scaler_path}")

X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# =============================================================
# âš™ï¸ Hybrid AIS + PSO Optimizer
# =============================================================

def fitness(params):
    """Train CNN-LSTM for given hyperparameters and return MSE."""
    lr, neurons, dropout = params
    model = Sequential([
        Conv1D(32, 2, activation='relu', input_shape=(X_train.shape[1], 1)),
        MaxPooling1D(),
        LSTM(int(neurons), activation='tanh'),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
    model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)
    pred = model.predict(X_test)
    return mean_squared_error(y_test, pred)


def hybrid_ais_pso(iterations=10, particles=6):
    """Hybrid Artificial Immune System + Particle Swarm Optimization."""
    lb = np.array([0.0005, 16, 0.1])  # learning_rate, neurons, dropout
    ub = np.array([0.01, 128, 0.5])
    dim = 3

    # Initialize PSO population
    pos = np.random.uniform(lb, ub, (particles, dim))
    vel = np.zeros((particles, dim))
    pbest = pos.copy()
    pbest_score = np.full(particles, np.inf)

    gbest = pos[0].copy()
    gbest_score = np.inf

    # PSO coefficients
    w, c1, c2 = 0.6, 1.5, 1.5

    # AIS parameters
    mutation_rate = 0.05

    for t in range(iterations):
        for i in range(particles):
            score = fitness(pos[i])
            if score < pbest_score[i]:
                pbest_score[i] = score
                pbest[i] = pos[i].copy()
            if score < gbest_score:
                gbest_score = score
                gbest = pos[i].copy()

        # Update velocities and positions (PSO)
        for i in range(particles):
            r1, r2 = np.random.rand(dim), np.random.rand(dim)
            vel[i] = w * vel[i] + c1 * r1 * (pbest[i] - pos[i]) + c2 * r2 * (gbest - pos[i])
            pos[i] = pos[i] + vel[i]

            # AIS mutation step
            mutation = np.random.normal(0, mutation_rate, dim)
            pos[i] += mutation

            # Boundary check
            pos[i] = np.clip(pos[i], lb, ub)

        print(f"[Iter {t+1}/{iterations}] Best MSE so far: {gbest_score:.6f}")

    return gbest, gbest_score


# =============================================================
# ðŸ§  Run Hybrid Optimization
# =============================================================
best_params, best_score = hybrid_ais_pso(iterations=6, particles=5)
lr, neurons, dropout = best_params
print(f"\n[INFO] âœ… Optimized Params -> lr={lr:.5f}, neurons={int(neurons)}, dropout={dropout:.2f}\n")

# =============================================================
# ðŸ§  Build Final CNN-LSTM Model
# =============================================================
model = Sequential([
    Conv1D(32, 2, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(),
    LSTM(int(neurons), activation='tanh'),
    Dropout(dropout),
    Dense(1)
])
model.compile(optimizer=Adam(learning_rate=lr), loss='mse', metrics=['mae'])

history = model.fit(X_train, y_train, epochs=50, batch_size=32,
                    validation_split=0.2, verbose=1)

model.save(model_path)
print(f"[INFO] Model saved at: {model_path}")

# =============================================================
# ðŸ“ˆ Predictions & Evaluation
# =============================================================
pred = model.predict(X_test)
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred)

metrics = {
    "MAE": float(mae),
    "MSE": float(mse),
    "R2": float(r2),
    "Best_Params": {
        "Learning_Rate": float(lr),
        "Neurons": int(neurons),
        "Dropout": float(dropout)
    }
}

with open(json_path, "w") as f: json.dump(metrics, f, indent=4)
with open(config_path, "w") as f: yaml.dump(metrics, f)
print(f"[INFO] Metrics saved to JSON and YAML.")

result_df = pd.DataFrame({"Actual": y_test, "Predicted": pred.flatten()})
result_df.to_csv(result_path, index=False)
print(f"[INFO] Result CSV saved to: {result_path}")

# =============================================================
# ðŸ“Š Visualization Section
# =============================================================

# 1ï¸âƒ£ Accuracy Graph
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("AIS+PSO AgroVolt Accuracy Graph")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(visual_dir, "ais_pso_agrovolt_accuracy_graph.png"))
plt.show()

# 2ï¸âƒ£ Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("AIS+PSO AgroVolt Feature Correlation Heatmap")
plt.savefig(os.path.join(visual_dir, "ais_pso_agrovolt_heatmap.png"))
plt.show()

# 3ï¸âƒ£ Comparison Scatter
plt.figure(figsize=(8,6))
plt.scatter(y_test, pred, color='green', alpha=0.6)
plt.title("AIS+PSO AgroVolt Predicted vs Actual")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.grid(True)
plt.savefig(os.path.join(visual_dir, "ais_pso_agrovolt_comparison_graph.png"))
plt.show()

# 4ï¸âƒ£ Result Line Graph
plt.figure(figsize=(10,6))
plt.plot(y_test[:200], label="Actual", linewidth=2)
plt.plot(pred[:200], label="Predicted", linestyle='dashed')
plt.title("AIS+PSO AgroVolt Result Graph (First 200 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Target Value")
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(visual_dir, "ais_pso_agrovolt_result_graph.png"))
plt.show()

print("\nâœ… AIS + PSO AGROVOLT MODEL EXECUTED SUCCESSFULLY âœ…")
print(f"[INFO] All outputs saved in: {base_path}")
```
