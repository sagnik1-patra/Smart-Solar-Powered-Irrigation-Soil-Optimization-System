```python
import os, json, yaml, warnings, joblib
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Dropout
from tensorflow.keras.optimizers import Adam
warnings.filterwarnings("ignore")

# =============================================================
# ðŸ“‚ PATHS
# =============================================================
base_path = r"C:\Users\NXTWAVE\Downloads\Smart Solar-Powered Irrigation & Soil Optimization System"
data_path = os.path.join(base_path, "archive", "spg.csv")
visual_dir = os.path.join(base_path, "visuals"); os.makedirs(visual_dir, exist_ok=True)

model_path  = os.path.join(base_path, "ais_ga_agrovolt_model.h5")
scaler_path = os.path.join(base_path, "ais_ga_agrovolt_scaler.pkl")
config_path = os.path.join(base_path, "ais_ga_agrovolt_config.yaml")
json_path   = os.path.join(base_path, "ais_ga_agrovolt_prediction.json")
result_path = os.path.join(base_path, "ais_ga_agrovolt_result.csv")

# =============================================================
# ðŸ“¥ Load Dataset
# =============================================================
print("[INFO] Loading dataset...")
df = pd.read_csv(data_path).dropna()
print(f"[INFO] Shape: {df.shape}")

X, y = df.iloc[:, :-1].values, df.iloc[:, -1].values
scaler = MinMaxScaler(); X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, scaler_path)

X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# =============================================================
# âš™ï¸ Hybrid AIS + GA Optimizer
# =============================================================

def fitness(params):
    """Train CNN-LSTM and return MSE."""
    lr, neurons, dropout = params
    model = Sequential([
        Conv1D(32, 2, activation='relu', input_shape=(X_train.shape[1],1)),
        MaxPooling1D(),
        LSTM(int(neurons), activation='tanh'),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer=Adam(learning_rate=lr), loss='mse')
    model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)
    pred = model.predict(X_test)
    return mean_squared_error(y_test, pred)

def hybrid_ais_ga(iterations=8, pop_size=6, mutation_rate=0.15, crossover_rate=0.7):
    lb, ub = np.array([0.0005,16,0.1]), np.array([0.01,128,0.5])
    dim = 3
    pop = np.random.uniform(lb, ub, (pop_size, dim))
    best_sol, best_score = None, np.inf

    for gen in range(iterations):
        scores = np.array([fitness(p) for p in pop])
        ranked = np.argsort(scores)
        memory = pop[ranked[:max(2,pop_size//2)]]  # AIS memory cells

        if scores[ranked[0]] < best_score:
            best_score, best_sol = scores[ranked[0]], pop[ranked[0]].copy()

        # === GA: Selection & Crossover ===
        new_pop = []
        for _ in range(pop_size//2):
            parents = memory[np.random.choice(memory.shape[0],2,replace=False)]
            if np.random.rand() < crossover_rate:
                cross_pt = np.random.randint(1, dim)
                child1 = np.concatenate([parents[0][:cross_pt], parents[1][cross_pt:]])
                child2 = np.concatenate([parents[1][:cross_pt], parents[0][cross_pt:]])
            else:
                child1, child2 = parents.copy()
            new_pop += [child1, child2]

        pop = np.array(new_pop)[:pop_size]

        # === AIS: Mutation / Clonal variation ===
        for i in range(pop_size):
            if np.random.rand() < mutation_rate:
                mutation = np.random.normal(0,0.05,dim)
                pop[i] += mutation
        pop = np.clip(pop, lb, ub)

        print(f"[Gen {gen+1}/{iterations}] Best MSE so far: {best_score:.6f}")

    return best_sol, best_score

# =============================================================
# ðŸ§  Run Optimization + Final Model
# =============================================================
best_params, best_score = hybrid_ais_ga(iterations=6, pop_size=6)
lr, neurons, dropout = best_params
print(f"\n[INFO] âœ… Optimized Params -> lr={lr:.5f}, neurons={int(neurons)}, dropout={dropout:.2f}\n")

# =============================================================
# ðŸ§  Final CNN-LSTM Model
# =============================================================
model = Sequential([
    Conv1D(32,2,activation='relu',input_shape=(X_train.shape[1],1)),
    MaxPooling1D(),
    LSTM(int(neurons),activation='tanh'),
    Dropout(dropout),
    Dense(1)
])
model.compile(optimizer=Adam(learning_rate=lr), loss='mse', metrics=['mae'])
history = model.fit(X_train,y_train,epochs=50,batch_size=32,validation_split=0.2,verbose=1)
model.save(model_path)

# =============================================================
# ðŸ“ˆ Predictions & Evaluation
# =============================================================
pred = model.predict(X_test)
mae, mse, r2 = mean_absolute_error(y_test,pred), mean_squared_error(y_test,pred), r2_score(y_test,pred)
metrics = {"MAE":float(mae),"MSE":float(mse),"R2":float(r2),
           "Best_Params":{"Learning_Rate":float(lr),"Neurons":int(neurons),"Dropout":float(dropout)}}

with open(json_path,"w") as f: json.dump(metrics,f,indent=4)
with open(config_path,"w") as f: yaml.dump(metrics,f)
pd.DataFrame({"Actual":y_test,"Predicted":pred.flatten()}).to_csv(result_path,index=False)

# =============================================================
# ðŸ“Š Visualization
# =============================================================
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'],label='Train Loss')
plt.plot(history.history['val_loss'],label='Validation Loss')
plt.legend(); plt.title("AIS+GA AgroVolt Accuracy Graph")
plt.xlabel("Epochs"); plt.ylabel("Loss"); plt.grid(True)
plt.savefig(os.path.join(visual_dir,"ais_ga_agrovolt_accuracy_graph.png")); plt.show()

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),cmap='coolwarm'); plt.title("AIS+GA AgroVolt Feature Correlation Heatmap")
plt.savefig(os.path.join(visual_dir,"ais_ga_agrovolt_heatmap.png")); plt.show()

plt.figure(figsize=(8,6))
plt.scatter(y_test,pred,color='green',alpha=0.6)
plt.title("AIS+GA AgroVolt Predicted vs Actual"); plt.xlabel("Actual"); plt.ylabel("Predicted")
plt.grid(True); plt.savefig(os.path.join(visual_dir,"ais_ga_agrovolt_comparison_graph.png")); plt.show()

plt.figure(figsize=(10,6))
plt.plot(y_test[:200],label="Actual",linewidth=2)
plt.plot(pred[:200],label="Predicted",linestyle='dashed')
plt.title("AIS+GA AgroVolt Result Graph (First 200 Samples)")
plt.xlabel("Sample Index"); plt.ylabel("Target Value")
plt.legend(); plt.grid(True)
plt.savefig(os.path.join(visual_dir,"ais_ga_agrovolt_result_graph.png")); plt.show()

print("\nâœ… AIS + GA AGROVOLT MODEL EXECUTED SUCCESSFULLY âœ…")
print(f"[INFO] All outputs saved in: {base_path}")
```
